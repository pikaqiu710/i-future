```python
# 计算每一个开通SIM卡号的历史账单费用总和
# iot_sample['sum'] = pd.DataFrame(iot_sample.loc[:,'账单费用201608':'账单费用201501'].sum(axis=1))
iot1.loc[:,'sum'] = iot1.loc[:,'账单费用201608':'账单费用201501'].sum(axis=1)
#iot2 = iot1.iloc[:,7:].sum(axis = 1) 
```

```python
# 选出账单费用为0的开通/申销卡号
print(iot2['状态'].value_counts())

iot2_开通 = iot2[iot2['状态']=='开通']
iot2_申销 = iot2[iot2['状态']=='申销']
```

```python
aggFuncs = {'开通时间': ['min','max'], 
            '客户名称': ['count'],
            '状态': {'开通': lambda x: x[x=='开通'].count(),
                     '申销': lambda x: x[x=='申销'].count(),
                     '欠销': lambda x: x[x=='欠销'].count(),
                     '无主': lambda x: x[x=='无主'].count(),
                     '待激活': lambda x: x[x=='待激活'].count(),
                     '停机': lambda x: x[x=='停机'].count()
                    }
            }
iot3 = pd.DataFrame(iot2.groupby(['客户名称']).agg(aggFuncs).sort_values(by=('客户名称','count'), ascending=False))
```
